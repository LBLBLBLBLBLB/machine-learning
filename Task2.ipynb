```python
import pandas as pd
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/emails_classification.csv')
data.head()
```

```python
y = data['Prediction'].values
X = data.drop('Prediction', axis=1).values
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.03, random_state=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(units=7, activation = 'relu', input_dim = X_train.shape[1],kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=7, activation = 'relu' ,kernel_initializer="normal"))
model.add(Dropout(rate = 0.3))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(optimizer='sgd', loss ='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=2,epochs=300,validation_data=(X_train,y_train), verbose=1)
```
